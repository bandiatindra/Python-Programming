```python
file = open('FloridaVoters.html','r') ###change your file path to where is 
data = file.readlines()
file.close()

print data
```

```python
list = []
for line in data: 
    if '<td>' in line: ###find any line with these characters
        list.append(line)
print list
```

```python
i=0
for strings in list:
    string = str(list[i])
    string = string.lstrip("<td>").rstrip("</td>\r\n")
    string = string.replace(',','')
    if (i)%6 != 0: ### if i divided by 6 does not = 0, bc every 6th value is county
        list[i]=int(string)
    else:
        list[i]=string
    i+=1
print list
```
